Numpy code is faster than python code. 
if we working with numPy data use numPy methods. If we working with python data types use python methods because going back to the start numPy has been optimized to perform numerical calculations. so whenever you want to perform numerical calculations always remember to use the numPy version of an aggregation function 

```python
# magic function which have % sign at the starting 
# %timeit - This is going to time how long a particular line of code takes to run 
# 8.16 ms = 8160 microseconds 

%timeit sum(massive_array) # Python's sum()
%timeit np.sum(massive_array) # NumPy's sum()
```

```python
# NumPy np.sum() us 168.25 times faster than python sum 
8160/48.5
```
